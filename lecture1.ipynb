# 6) EDA – Exploratory Data Analysis (Keşifsel Veri Analizi)

## 6.1. Tek Değişkenli Analiz (Univariate)

### Sıcaklık dağılımı

```python
plt.hist(df["temp"])
plt.title("Sıcaklık Dağılımı")
plt.xlabel("Sıcaklık (C)")
plt.ylabel("Frekans")
plt.show()
# Histogram → bir değişkenin hangi aralıklarda yoğunlaştığını gösterir.
```

### Nem dağılımı

```python
sns.histplot(df["RH"], kde=True)
```

```python
# Yorum:
# sağa çarpık sola çarpık nedir??????

# Yangın riskini yorumlarken önemlidir.
```

## 6.2. Kategorik Veri Analizi

### Ay bazında yangın sayısı

```python
sns.countplot(x="month", data=df)
```

### Gün bazında yangın sayısı

```python
sns.countplot(x="day", data=df)
```

## 6.3. Çift Değişkenli Analiz (Bivariate)

### Sıcaklık – Yanan alan ilişkisi

```python
plt.scatter(df["temp"], df["area"])
plt.xlabel("Sıcaklık")
plt.ylabel("Yanan Alan")
plt.title("Sıcaklık vs Yanan Alan")
plt.show()
```

## 6.4. Korelasyon Analizi

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
# Korelasyon = iki değişken arasındaki ilişki

# +1 → pozitif ilişki

# -1 → negatif ilişki

# 0 → ilişki yok

# Isı haritası → ilişkileri renklerle gösterir.

# 📌 Veri seti yorumu:

# FFMC, DMC, DC birbiri ile yüksek korelasyonlu

# area ile korelasyon zayıf → bu veri seti zor bir regresyon problemidir
```
